```python
s = pd.read_csv("data.csv")

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Asumiendo que ya tienes el DataFrame cargado como 'df'
# Si no, descomenta la siguiente línea y ajusta la ruta del archivo
# df = pd.read_csv('ruta_a_tu_archivo.csv')

# Calcular la media de Speechiness por año
loudness_70s = s[s["year"] >= 1985]
loudness_70s = loudness_70s[loudness_70s["year"] < 2020]

loudness_70s = loudness_70s.groupby('year')['loudness'].mean().reset_index()
#loudness_by_year = s.groupby('year')['loudness'].mean().reset_index()

# Visualizar la tendencia de Speechiness a lo largo del tiempo
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='loudness', data=loudness_70s)
plt.title('Promedio de Loudness por Año')
plt.ylabel('Loudness')
plt.xlabel('Año')
plt.show()


loudness_70s.describe()
```
